```python
s = [1,2,3,4,5,6,7,8,9,1,1,1,2,1,3,1,4,1,5,1,6,1,7,1,8,1,9]
ind = 0
ind1 = 0

def printv():  #IGROk(ввод)
    print("Введите первые две цифры нули, если хотите добавить числа")
    x1 = int(input('Введите x: '))
    x2 = int(input('Введите y: '))
    ind = x1-1 + (x2-1) * 9
    y1 = int(input('Введите x: '))
    y2 = int(input('Введите y: '))
    ind1 = y1-1 + (y2-1) * 9
    return ind, ind1

def deleteNumerals(ind, ind1): #мы перебираем все элементы , проверяем равно ли оно нулю , если да то добавляем 1
    if ind > ind1:
        ind, ind1 = ind1, ind
    i = ind + 1

    while i < ind1:
        if s[i] != 0:
            break
        i += 1

    if s[ind] != 0 and ind != ind1 and i == ind1 and (s[ind] + s[ind1] == 10 or s[ind] == s[ind1]):
        s[ind], s[ind1] = 0, 0
        return True

    i = ind + 9
    while i < ind1:
        if s[i] != 0:
            break
        i += 1

    if s[ind] != 0 and ind != ind1 and i == ind1 and (s[ind] + s[ind1] == 10 or s[ind] == s[ind1]):
        s[ind], s[ind1] = 0, 0
        return True
    return False

def addNumerals(): # функция добовления строк
    for i in range(0, len(s)):
        if s[i] != 0:
            s.append(s[i])

def printS():    #функция вывода
    numOfLines = (len(s) + 8) / 9

    numOfNumeralsInNumOfLines = 0
    temp = numOfLines
    while temp >= 1:
        numOfNumeralsInNumOfLines += 1
        temp /= 10

    for i in range(numOfNumeralsInNumOfLines + 1):
        print(end = ' ')

    for i in range(1, 10):
        print(i, end = ' ')
    print()

    for i in range(numOfNumeralsInNumOfLines + 1):
        print(end = ' ')

    for i in range(1, 10):
        print('-', end = ' ')
    print()

    i = 0
    currentLine = 1
    while i < len(s):
        print(str(currentLine).zfill(numOfNumeralsInNumOfLines), end="|")
        for k in range(0, 9):
            if i < len(s):
                print(s[i],end=" ")
                i += 1
        print()
        currentLine += 1
    print()

def prov(): # проверка на окнчание игры
    for j in s:
        if s[j] != 0:
            return True
    return False

while prov(): # вводим числа ,если индекс меньше 0 то мы вносим элементы , иначе удаляем
    printS()
    ind, ind1 = printv()
    if ind1 < 0:
        addNumerals()
    else:
        deleteNumerals(ind, ind1)
```
